# Python code to generate a QR code for the Digikala website


```python
# Requirement
# First, install qrcode package
!pip install qrcode
```

```python
import qrcode
from qrcode.image.styledpil import StyledPilImage
from qrcode.image.styles.moduledrawers.pil import RoundedModuleDrawer
from qrcode.image.styles.colormasks import RadialGradiantColorMask

# URL for Digikala
url = "https://www.digikala.com"

# Create QR code instance
qr = qrcode.QRCode(
    version=1,
    error_correction=qrcode.constants.ERROR_CORRECT_H,
    box_size=10,
    border=4,
)

# Add data
qr.add_data(url)
qr.make(fit=True)

# Create and save QR code with custom styling
img = qr.make_image(
    image_factory=StyledPilImage,
    module_drawer=RoundedModuleDrawer(),
    color_mask=RadialGradiantColorMask(
        center_color=(0, 150, 136),  # Teal color
        edge_color=(255, 87, 34)     # Orange color (matching Digikala's brand)
    ),
    back_color="white"
)

# Save the QR code
img.save("digikala_qr.png")
print("✅ QR code for Digikala has been saved as 'digikala_qr.png'")

# Display QR code (if running in an environment that supports it)
img.show()
```
